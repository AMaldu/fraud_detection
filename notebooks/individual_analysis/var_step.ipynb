```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from src.utils import pie_plot
import numpy as np
```

```python
data = pd.read_csv('../../data/processed/fraud_viz_df.csv') 
```

```python
data
```

```python
data['day_of_month'] = np.ceil(data['step'] / 24).astype(int)
data
```

```python
data['day_of_month'].value_counts().sort_index()
```

```python
# plot by week 
plt.figure(figsize=(12, 6)) 
sns.kdeplot(data=data['day_of_month'])
plt.xlabel('Day of Month')
plt.ylabel('Density')
plt.title('Density Plot of Day of Month')
plt.show()
```

```python
data['hour'] = (data['step'] - 1) % 24
```

```python
data
```

```python
# viz avg per day 
plt.figure(figsize=(14, 6)) 
sns.kdeplot(data=data['hour'])
plt.xlabel('Hour of the Day')
plt.ylabel('Density')
plt.title('Density Plot of Hour of the Day')
plt.show()
```
